```python
import getml
from getml import data
import zipfile
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
getml.engine.launch()
getml.set_project("danish")
```

```python
temp_path = './data/temp'
roles = {'target': ['value'], 'time_stamp': ['date']}
# Extract zip file to the temporary directory
with zipfile.ZipFile('./data/time_series.zip', 'r') as zip_ref:
    zip_ref.extractall(temp_path)
df = data.DataFrame.from_csv('data/temp/danish_atm_daily/danish_atm_daily_0.csv',
                             name = 'danish',
                             roles=roles,
                             ignore=True)
```

```python
df_p = pd.read_csv('data/temp/danish_atm_daily/danish_atm_daily_0.csv')
```

```python
df_p.tail(60)
```

```python
split = getml.data.split.time(df, "date", test=getml.data.time.datetime(2017, 2, 11))
```

```python
time_series = getml.data.TimeSeries(
    population=df,
    time_stamps="date",
    split=split,
    horizon=getml.data.time.hours(1),
    memory=getml.data.time.days(7),
    lagged_targets=True,
)

time_series
```

```python
type(time_series.train)
```

```python
seasonal = getml.preprocessors.Seasonal()

fast_prop = getml.feature_learning.FastProp(
    loss_function=getml.feature_learning.loss_functions.SquareLoss,
    num_threads=1,    
    num_features=20,
)

predictor = getml.predictors.XGBoostRegressor()
```

```python
pipe = getml.pipeline.Pipeline(
    tags=["memory: 7d", "horizon: 1h", "fast_prop"],
    data_model=time_series.data_model,
    preprocessors=[seasonal],
    feature_learners=[fast_prop],
    predictors=[predictor],
)
```

```python
df[split == "test"]
```

```python
pipe.fit(time_series.train)
```

```python
pipe.score(time_series.test)
```

```python
col_data = "black"
col_getml = "darkviolet"
```

```python
pipe.features._load_features
```

```python
names, correlations = pipe.features.correlations()

plt.subplots(figsize=(20, 10))

plt.bar(names, correlations, color=col_getml)
plt.title("Feature Correlations")
plt.xlabel("Features")
plt.ylabel("Correlations")
plt.xticks(rotation="vertical")
plt.show()
```

```python
names, importances = pipe.features.importances()

plt.subplots(figsize=(20, 10))

plt.bar(names, importances, color=col_getml)
plt.title("Feature Importances")
plt.xlabel("Features")
plt.ylabel("Importances")
plt.xticks(rotation="vertical")
plt.show()
```
